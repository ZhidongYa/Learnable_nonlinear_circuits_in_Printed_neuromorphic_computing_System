```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize as op
import math
import torch
```

```python
# fitting
def f(x, eta1,eta2,eta3,eta4):
    return -(eta1 +eta2*np.tanh((x-eta3)*eta4))
```

```python
data1 = pandas.read_csv("inv_output.csv")
```

```python
def MSE(y, t):
    return  np.sum((y - t)**2)/len(y)
```

```python
validresult = []
x=np.linspace(-2,2,4001)
for k in range(9000,9999):
    f1 = data1.iloc[k, :]
    y = np.array([float(f2) for f2 in f1.to_numpy()]) 
    eta1,eta2,eta3,eta4= op.curve_fit(f, x, y, maxfev=1000000,p0=[0.,1.,0.,1.], bounds=([-np.inf, 0, -np.inf, 0],[np.inf, np.inf, np.inf, np.inf]))[0]
    ys = -(eta1 +eta2*np.tanh((x-eta3)*eta4))
    error2 = MSE(y,ys)
    if error2 > 0.0011904407837462858:
        validresult.append(k)
```
