```python
# packages
import os
import time
import math
import torch
import random
import pickle
import pandas
import calendar
import training
import config
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import TensorDataset, DataLoader
os.environ['KMP_DUPLICATE_LIB_OK']='True'
torch.set_default_tensor_type(torch.FloatTensor)
```

```python
input_features = config.topology[0]
if input_features == 11: name = 'invWithRatio_dataset.txt'
if input_features == 7: name = 'inv_dataset.txt'
if input_features == 8: name = 'invWithR1234_dataset.txt'  
print(name)

with open(name, 'rb') as file:
    dataset = pickle.load(file)
y = dataset['Y']
X = dataset['X']
x_train, y_train, x_valid, y_valid,x_test,y_test = training.SplitData(X,y,0.6,0.2,seed=0)
t = X.max(dim= 0)[0]
tmin= X.min(dim= 0)[0]
X1 = (X-tmin)/(t-tmin)
x_train, y_train, x_valid, y_valid,x_test,y_test = training.SplitData(X1,y,0.6,0.2,seed=0)
train_ds = TensorDataset(x_train, y_train)
train_loader = DataLoader(train_ds, batch_size = y_train.shape[0])
valid_ds = TensorDataset(x_valid, y_valid)
valid_loader = DataLoader(valid_ds, batch_size = y_valid.shape[0])
test_ds = TensorDataset(x_test, y_test)
test_loader = DataLoader(test_ds, batch_size = y_test.shape[0])
```
