```python
# net level
import pNN_NetLevel_LNC_variation1 as pNN

# packages
import os
import math
import torch
import random
import pickle
import calendar
import numpy as np
import config
import training_together_variation as training 
import matplotlib.pyplot as plt
from torch.utils.data import TensorDataset, DataLoader
os.environ['KMP_DUPLICATE_LIB_OK']='True'
torch.set_default_tensor_type(torch.FloatTensor)
```

```python
datasets = os.listdir('./dataset')
datasets = [f for f in datasets if (f.startswith('Dataset') and f.endswith('.p'))] 
#datasets
```

```python

result = torch.zeros((13, 2, 10))
epsilons = [0.05]
for i in range(len(datasets)):
    dataset = datasets[i]
    datapath = os.path.join(f'./dataset/{dataset}')
    with open(datapath, 'rb') as f:
        data = pickle.load(f)
    X_train    = data['X_train']
    y_train    = data['y_train']
    X_valid    = data['X_valid']
    y_valid    = data['y_valid']
    X_test     = data['X_test']
    y_test     = data['y_test']
    data_name  = data['name']

    N_class    = data['n_class']
    N_feature  = data['n_feature']
    N_train    = X_train.shape[0]
    N_valid    = X_valid.shape[0]
    N_test     = X_test.shape[0]
    
    # generate tensordataset
    trainset = TensorDataset(X_train, y_train)
    validset = TensorDataset(X_valid, y_valid)
    testset  = TensorDataset(X_test, y_test)

    # batch
    train_loader = DataLoader(trainset, batch_size=N_train)
    valid_loader = DataLoader(validset, batch_size=N_valid)
    test_loader  = DataLoader(testset,  batch_size=N_test)
    
    for j in range(len(epsilons)):
        epsilon = epsilons[j]
        
        for n in range(10):
            seed = n      
            setup = f'dataset_{data_name}_lr_0.001_epsilon_{epsilon}_seed_{seed}_variation'
            print(setup)
            
            if os.path.exists(f'./results_variation/pNN_{setup}'):
                print('File exists, pass.')
            else:
                torch.manual_seed(seed)
                model = pNN.pNN([N_feature, 3, N_class], config.N, epsilon)
    
                optimizer = torch.optim.Adam([{'params':model.GetParam('theta_'),'lr':1e-1},
                                  {'params':model.GetParam('eta_'), 'lr':0.001},
                                  {'params':model.GetParam('inv_'),'lr':0.001}])
    
    
                NN, acc_valid = training.training_pNN(model, train_loader, valid_loader, optimizer, pNN.lossfunction)
                result[i,j,n] = acc_valid
                torch.save(NN, f'./results_variation/pNN_{setup}')

#a = {'acc_valid': result}
#with open('acc_valid_variation.txt', 'wb') as file:
    #pickle.dump(a, file)
```

## valid_loss

```python
acc_loss = torch.zeros((13, 2, 10))
lrs = [0.05, 0.1]
for i in range(len(datasets)):
    dataset = datasets[i]
    datapath = os.path.join(f'./dataset/{dataset}')
    with open(datapath, 'rb') as f:
        data = pickle.load(f)
    X_train    = data['X_train']
    y_train    = data['y_train']
    X_valid    = data['X_valid']
    y_valid    = data['y_valid']
    X_test     = data['X_test']
    y_test     = data['y_test']
    data_name  = data['name']

    N_class    = data['n_class']
    N_feature  = data['n_feature']
    N_train    = X_train.shape[0]
    N_valid    = X_valid.shape[0]
    N_test     = X_test.shape[0]
    
    # generate tensordataset
    trainset = TensorDataset(X_train, y_train)
    validset = TensorDataset(X_valid, y_valid)
    testset  = TensorDataset(X_test, y_test)

    # batch
    train_loader = DataLoader(trainset, batch_size=N_train)
    valid_loader = DataLoader(validset, batch_size=N_valid)
    test_loader  = DataLoader(testset,  batch_size=N_test)
    
    for j in range(len(lrs)):
        epsilon = lrs[j]
        
        for n in range(10):
            seed = n      
            if epsilon == 0.05:
                setup = f'dataset_{data_name}_seed_{seed}_epsilon_0.05_nonlearnable_variation'
            else:
                setup = f'dataset_{data_name}_seed_{seed}_epsilon_0.10_nonlearnable_variation'
            #setup = f'dataset_{data_name}_epsilon_{epsilon}_seed_{seed}_variation'
            model = torch.load(f'./results_variation/pNN_{setup}')
            
            
            with torch.no_grad():
                for X_valid, y_valid in valid_loader:
                
                    prediction_valid = model(X_valid)
                    acc_valid = pNN.lossfunction(prediction_valid, y_valid)
                
                    acc_loss[i,j,n] = acc_valid
acc_loss
```
